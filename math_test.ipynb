```python
import numpy as np
from math import sqrt
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p1 = (1, 2)
p2 = (3, 4)
p3 = (4, 5)
p4 = (2, 3)

points = [p1, p2, p3, p4]
```

```python
def slope_intercept(p1, p2):
    slope = (p2[1] - p1[1]) / (p2[0] - p1[0])
    return slope, p1[1] - slope * p1[0]

def find_intercept(p1, p2, p3, p4):
    slope1, inter1 = slope_intercept(p1, p2)
    slope2, inter2 = slope_intercept(p3, p4)
    try:
        x = (inter1 - inter2) / (slope2 - slope1)
        y = slope1 * x + inter1
        return (x, y)
    except ZeroDivisionError:
        return False

find_intercept(p1, p2, p3, p4)
sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
```

```python
for i in range(0, len(points), 2):
    plt.plot([points[i][0], points[i+1][0]], [points[i][1], points[i+1][1]], 'go-')
plt.grid('on')
plt.show()
```
